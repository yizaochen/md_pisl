### Part 1: Get collocation points and integration kernel for whole domain

```julia
Nh = 64    # The number of Spectral element
Np = 4     # The order of polynomial which used to interpolate and integration

# Define Physical Domain
xratio = 180 # unit: angstrom
xavg = 0   # unit: angstrom

x, w, Ldx, L = getLagrange(Np, xratio/Nh)
e_norm = x[end] - x[1]
interpo_xs = x .+ x[end]

N, xref, w0, Ldx, w = get_fem_xref_weights_basis(Nh, Np, xratio, xavg);
```

```julia
function symmetry_wall_potential(left_x::Float64, right_x::Float64, sigma::Float64, scalefactor::Float64, xvalue::Float64)
    if xvalue < 0.
        return scalefactor * gaussian(xvalue, left_x, sigma)
    else
        return scalefactor * gaussian(xvalue, right_x, sigma)
    end
end
```

```julia
potential = symmetry_wall_potential.(-180., 180., 5., 50., xref);
```

```julia
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(4,4))

ax.plot(xref, potential, linewidth=2)
ax.set_xlabel("\$\\phi\$", fontsize=12)
ax.set_ylabel("\$V(x)\$ (kcal/mol)", fontsize=12)
#ax.axvline(-170, color="grey", alpha=0.5)
#ax.axvline(170, color="grey", alpha=0.5)
#ax.set_yticks(0:5:20)
ax.set_xticks(-180:90:180)
#ax.tick_params(axis="both", labelsize=8)
#ax.set_title("Model 1", fontsize=12)

plt.tight_layout()
#plt.savefig("symmetry_wall_phi.pdf")
plt.show()
```

### Part 2: Distance-dependent diffusion coefficient $D(x)$

```julia
sigmoid(z::Real) = one(z) / (one(z) + exp(-z));
```

```julia
xarray = range(-5,stop=5,length=200)
D_array = sigmoid.(xarray) .* 20;
```

```julia
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(8,4))

ax.plot(xarray, D_array)
ax.set_xlabel("x")
ax.set_ylabel("D")

plt.tight_layout()
#plt.savefig("distance_dependent_D.pdf")
plt.show()
```

### Part 3: Langenvin dynamics simulations

```julia
# physical parameters
save_freq = 0.01 # unit: s, Save trajectory per 100 ps
total_times = 100 # unit: s, totol simulate 10 us, 10e-6
dt = 0.001 # unit: s  # Integration Times 1ps
#kbT = 1
#D = 4.845e9; # Diffusion coefficient, unit: angstrom^2 s-1
tau = Int(round(total_times / save_freq, digits=0))  # Number of photons
n_interval = Int(round(save_freq / dt, digits=0));

time_info = @sprintf "Save trajectory per %d timesteps. There will be %d data points." n_interval tau
println(time_info)
```

```julia
# Initialize Variables
y_record = zeros(tau+1, 1)  # In HMM, observing variables Y(t)
t_record = zeros(tau+1, 1)

ypos = 0 # Initial Position, from mean position
t = 0

# Simulation Start
for tau_id = 1:tau
    y_record[tau_id] = ypos
    t_record[tau_id] = t
    for i = 1:n_interval
        D = sigmoid.(ypos) .* 20 
        F = 0
        ypos = ypos + (dt*2*D)^(1/2) * randn()
        t += dt   
    end
end
y_record[tau+1] = ypos
t_record[tau+1] = t;
```

```julia
lbfz = 14
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(10,4))

ax.plot(t_record, y_record)
ax.set_ylabel("x", fontsize=lbfz)
ax.set_xlabel("Time", fontsize=lbfz)
ax.set_ylim(-9.22752974472737, 97.05074068056088)

plt.tight_layout()
#plt.savefig("distance_dependent_D_traj1.png", dpi=150)
plt.show()
```

```julia
# Initialize Variables
y_record1 = zeros(tau+1, 1)  # In HMM, observing variables Y(t)
t_record1 = zeros(tau+1, 1)

ypos = 0 # Initial Position, from mean position
t = 0

# Simulation Start
for tau_id = 1:tau
    y_record1[tau_id] = ypos
    t_record1[tau_id] = t
    for i = 1:n_interval
        D = sigmoid.(ypos) .* 20 
        F = 0
        ypos = ypos + (dt*2*D)^(1/2) * randn()
        t += dt   
    end
end
y_record1[tau+1] = ypos
t_record1[tau+1] = t;
```

```julia
lbfz = 14
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(10,4), sharey=true)

ax.plot(t_record, y_record, label="Trajectory 1", alpha=0.8)
ax.plot(t_record1, y_record1, label="Trajectory 2", alpha=0.8)
ax.set_ylabel("x", fontsize=lbfz)
ax.set_xlabel("Time", fontsize=lbfz)
ax.legend()

plt.tight_layout()
#plt.savefig("distance_dependent_D_traj.png", dpi=150)
plt.show()
```
