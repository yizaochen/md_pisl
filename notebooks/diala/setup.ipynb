### Part 2: Run pdb2gmx on Allostery

```python
host = 'alad'
```

```python
print('cd /home/yizaochen/simulation/diala') 
```

```python
cmd = f'gmx pdb2gmx -f {host}.pdb -o {host}.gro -water tip3p'
print(cmd)
# Choose "CHARMM36 all-atom force field (March 2017)"
```

### Part 3: Set the water box and solvation

```python
cmd = f'gmx editconf -f {host}.gro -o {host}_dode.gro -c -d 1.0 -bt dodecahedron'
print(cmd)

cmd = f'gmx solvate -cp {host}_dode.gro -cs spc216.gro -o {host}_solv.gro -p topol.top'
print(cmd)
```

### Part 4: Adding Ions

```python
cmd = f'gmx grompp -f mdp/ions.mdp -c {host}_solv.gro -p topol.top -o ions.tpr'
print(cmd)

cmd = f'gmx genion -s ions.tpr -o {host}_solv_ions.gro -p topol.top -pname K  -neutral -conc 0.15'
print(cmd)
```

### Part 5: Minimization - Steep Descent

```python
cmd = f'gmx grompp -f mdp/minim.mdp -c {host}_solv_ions.gro -p topol.top -o em1.tpr'
print(cmd)

cmd = f'gmx mdrun -v -deffnm em1'
print(cmd)
```

```python
cmd = f'gmx energy -f em1.edr -o mini_sd_energy.xvg'
print(cmd)
```

### Part 6: Make index file

```python
cmd = f'gmx make_ndx -f em1.gro -o {host}0.ndx'
print(cmd)
```
