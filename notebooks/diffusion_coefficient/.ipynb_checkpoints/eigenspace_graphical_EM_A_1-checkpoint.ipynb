### Part 1: Get collocation points and integration kernel for whole domain

```julia
Nh = 64    # The number of Spectral element
Np = 4     # The order of polynomial which used to interpolate and integration
Nv = 72    # Number of eigenvectors

# Define Physical Domain
xratio = 64. # unit: angstrom
xavg = 0.   # unit: angstrom

x, w, Ldx, L = getLagrange(Np, xratio/Nh)
e_norm = x[end] - x[1]
interpo_xs = x .+ x[end]

N, xref, w0, Ldx, w = get_fem_xref_weights_basis(Nh, Np, xratio, xavg);
```

### Part 2: Set $V(x)$ and $p_{\rm eq}(x)$

```julia
sigma_kref = 8.
kref = get_k_by_sigma(sigma_kref)
Vref = harmonic_well_k_mean(xref, kref, xavg)
rho_eq = get_rhoeq(Vref, w0)
p_eq = rho_eq .* rho_eq;
```

$p(\textbf{x}_{n-1}, \textbf{x}_{n}|\textbf{Y})$

```julia
p_x_prev_marginal_vec_given_Y = zeros(N,1)
for idx = 1:N
    p_x_prev_marginal_vec_given_Y[idx] = sum(w0 .* xi_xspace[idx,:]) # \int_x5 p(x4,x5) dx5 = p(x4)
end
sum(w0 .* p_x_prev_marginal_vec_given_Y)
```

$p(\textbf{x}_{n-1}|\textbf{Y}) = \int p(\textbf{x}_{n-1}, \textbf{x}_{n}|\textbf{Y}) d\textbf{x}_{n}$  
$p(\textbf{x}_{n}|\textbf{Y}) = \int p(\textbf{x}_{n-1}, \textbf{x}_{n}|\textbf{Y}) d\textbf{x}_{n-1}$ 

```julia
p_x_prev_given_Y = zeros(N,1)
p_x_i_given_Y = zeros(N,1)
for idx = 1:N
    p_x_prev_given_Y[idx] = sum(w0 .* xi_xspace[idx,:]) # \int_x5 p(x4,x5) dx5 = p(x4)
    p_x_i_given_Y[idx] = sum(w0 .* xi_xspace[:, idx])
end
println(sum(w0 .* p_x_prev_given_Y))
println(sum(w0 .* p_x_i_given_Y))
```

$\xi(\textbf{s}_{n-1}, \textbf{s}_{n})$

```julia
# Check the sum of matrix is 1
sum(xi_eigenspace)
```

$p(\textbf{s}_{n-1}|\textbf{Y}) = \sum_{\textbf{x}_{n}} p(\textbf{s}_{n-1}, \textbf{x}_{n}|\textbf{Y})$  
$p(\textbf{s}_{n}|\textbf{Y}) = \sum_{\textbf{x}_{n-1}} p(\textbf{s}_{n-1}, \textbf{x}_{n}|\textbf{Y})$ 

```julia
p_s_prev_given_Y = zeros(Nv,1)
p_s_i_given_Y = zeros(Nv,1)
for idx = 1:Nv
    p_s_prev_given_Y[idx] = sum(xi_eigenspace[idx,:])
    p_s_i_given_Y[idx] = sum(xi_eigenspace[:, idx])
end
println(sum(p_s_prev_given_Y))
println(sum(p_s_i_given_Y))
```

### Part 11-2: EM Transition Matrix A

```julia
D_em_array = zeros(Nv)
log_A_array = zeros(Nv)
lambdai_deltat_array = zeros(Nv)
for eigv_idx = 1:Nv
    numerator = sum([xi_store_mat[eigv_idx,1,photon_idx] for photon_idx=2:tau])
    denominator = 0.
    for l=1:Nv
        denominator += sum([xi_store_mat[eigv_idx,l,photon_idx] for photon_idx=2:tau])
    end
    A = numerator / denominator
    #D_em_array[eigv_idx] = log(1-A) / (-2 * eigenvalues_prime[eigv_idx] * save_freq)
    D_em_array[eigv_idx] = log(1-A) / (-2 * 1 * save_freq)
    log_A_array[eigv_idx] = log(1-A)
    lambdai_deltat_array[eigv_idx] = -2 * eigenvalues_prime[eigv_idx] * save_freq
end
```

### Part 12: EM Iteration

```julia
n_iter = 20
D_em_mat = zeros(n_iter, Nv);
```

```julia
D_guess = 500.

for iter_id=1:n_iter
    println(iter_id)
    alpha_mat, beta_mat, Anorm_vec, r_sqrt_array = Diffusity.forward_backward_v0(Nv, tau, w0, p_eq, N, eigenvect_mat_prime, y_record, 
        xref, e_norm, interpo_xs, Np, k_delta, D_guess, eigenvalues_prime, save_freq)
    
    xi_store_mat = zeros(Nv,Nv,tau)
    for photon_idx=2:tau
        xi_store_mat[:,:,photon_idx] = Diffusity.get_xi_s_space(D_guess, eigenvalues_prime, save_freq, alpha_mat, photon_idx, 
        Nv, y_record, xref, e_norm, interpo_xs, Np, w0, k_delta, eigenvect_mat_prime, beta_mat)
    end
    
    D_em_array = zeros(Nv)
    for eigv_idx = 1:Nv
        numerator = sum([xi_store_mat[eigv_idx,eigv_idx,photon_idx] for photon_idx=2:tau])
        denominator = 0.
        for l=1:Nv
            denominator += sum([xi_store_mat[eigv_idx,l,photon_idx] for photon_idx=2:tau])
        end
        A = numerator / denominator
        D_em_array[eigv_idx] = log(A) / (-2 * eigenvalues_prime[eigv_idx] * save_freq)
    end
    D_em_mat[iter_id,:] = D_em_array
    D_guess = D_em_array[2]
end
```
