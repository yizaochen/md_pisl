### Part 1: Get collocation points and integration kernel for whole domain

```julia
Nh = 64    # The number of Spectral element
Np = 4     # The order of polynomial which used to interpolate and integration
Nv = 72    # Number of eigenvectors

# Define Physical Domain
xratio = 64. # unit: angstrom
xavg = 0.   # unit: angstrom

x, w, Ldx, L = getLagrange(Np, xratio/Nh)
e_norm = x[end] - x[1]
interpo_xs = x .+ x[end]

N, xref, w0, Ldx, w = get_fem_xref_weights_basis(Nh, Np, xratio, xavg);
```

### Part 2: Set $V(x)$ and $p_{\rm eq}(x)$

```julia
sigma_kref = 8.
kref = get_k_by_sigma(sigma_kref)
Vref = harmonic_well_k_mean(xref, kref, xavg)
rho_eq = get_rhoeq(Vref, w0)
p_eq = rho_eq .* rho_eq;
```

```julia
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(8,4))

ax = axes[1]
ax.plot(xref, Vref, linewidth=2)
ax.set_xlabel("\$x\$(L)", fontsize=12)
ax.set_ylabel("\$V(x)\$", fontsize=12)
ax.set_xticks(-64:16:64)
#ax.set_ylim(0, 10)

ax = axes[2]
ax.plot(xref, p_eq, linewidth=2)
ax.set_xlabel("\$x\$(L)", fontsize=12)
ax.set_ylabel("\$p_{\\rm eq}(x)\$", fontsize=12)
ax.set_xticks(-64:16:64)
ax.set_title(@sprintf "\$\\sigma_{\\rm harmonic} = %.1f\$ L" sigma_kref)

plt.tight_layout()
#plt.savefig("/home/yizaochen/Desktop/drawzone_temp/0/harmonic_well_r.svg")
plt.show()
```

### Part 9: Numerical and Analytical $\frac{d \ln{P(Y)}}{dD}$

### Part 10: $\left< \hat{\alpha}_{t-1} | \psi_i \right> \left< \psi_i | \textbf{y}_t | \hat{\beta}_{t} \right>$

```julia
big_photon_mat = get_big_photon_mat(N, Nv, w0, k_delta, xref, eigenvect_mat_prime)
idx_array = [find_nearest_point(y_record[time_idx], xref, e_norm, interpo_xs, Np) for time_idx=1:tau]
alpha_mat, y_beta_mat = Diffusity.get_alpha_beta_mat(Nv, tau, w0, p_eq, N, eigenvect_mat_prime, D_old, eigenvalues_prime, save_freq, big_photon_mat, idx_array);
```

```julia
fig, axes = plt.subplots(ncols=3, nrows=2, figsize=(12,6), sharey=true)

xarray = 1:72
xticks = 1:5:72
time_idx_array = [5, 500, 2500, 5000, 8000, 9998]

idx = 1
for row_id=1:2
    for col_id=1:3
        ax = axes[row_id, col_id]
        time_idx = time_idx_array[idx]
        alpha_beta = alpha_mat[:, time_idx-1] .* y_beta_mat[:, time_idx]
        ax.plot(xarray, alpha_beta, "b.", label=@sprintf "t=%d" time_idx)
        ax.vlines(xarray, ymin=0, ymax=alpha_beta)
        ax.set_xticks(xticks)
        ax.legend()
        if col_id == 1
            ax.set_ylabel("\$\\left< \\hat{\\alpha}_{t-1} | \\psi_i \\right> \\left< \\psi_i | \\mathbf{y}_t | \\hat{\\beta}_{t} \\right>\$")
        end
        idx += 1
    end
end

plt.tight_layout()
#plt.savefig("/home/ytcdata/em_theory/Pictures/ch2/alpha_beta_in_dldD.pdf")
plt.show()
```

### Part 11: $e^{-D \lambda'_i \Delta t} \left< \hat{\alpha}_{t-1} | \psi_i \right> \left< \psi_i | \textbf{y}_t | \hat{\beta}_{t} \right>$

```julia
D_now = 500
expLQDT = exp.(-(D_now .* eigenvalues_prime) .* save_freq);
```

```julia
fig, axes = plt.subplots(ncols=3, nrows=2, figsize=(12,6), sharey=true)

xarray = 1:72
xticks = 1:5:72
time_idx_array = [5, 500, 2500, 5000, 8000, 9998]

idx = 1
for row_id=1:2
    for col_id=1:3
        ax = axes[row_id, col_id]
        time_idx = time_idx_array[idx]
        alpha_beta = (expLQDT .* alpha_mat[:, time_idx-1]) .* y_beta_mat[:, time_idx]
        ax.plot(xarray, alpha_beta, "b.", label=@sprintf "t=%d" time_idx)
        ax.vlines(xarray, ymin=0, ymax=alpha_beta, label=@sprintf "sum=%.4f" sum(alpha_beta))
        ax.set_xticks(xticks)
        ax.legend()
        if col_id == 1
            ax.set_ylabel("\$e^{-D \\lambda'_i \\Delta t} \\left< \\hat{\\alpha}_{t-1} | \\psi_i \\right> \\left< \\psi_i | \\mathbf{y}_t | \\hat{\\beta}_{t} \\right>\$")
        end
        idx += 1
    end
end

plt.tight_layout()
#plt.savefig("/home/ytcdata/em_theory/Pictures/ch2/edt_alpha_beta_in_dldD.pdf")
plt.show()
```

### Part 12: $-\lambda'_i \Delta te^{-D \lambda'_i \Delta t} \left< \hat{\alpha}_{t-1} | \psi_i \right> \left< \psi_i | \textbf{y}_t | \hat{\beta}_{t} \right>$

```julia
LQDTexpLQDT =  -(eigenvalues_prime .* save_freq) .* exp.(-(D_now .* eigenvalues_prime) .* save_freq);
```

```julia
fig, axes = plt.subplots(ncols=3, nrows=2, figsize=(12,6), sharey=true)

xarray = 1:72
xticks = 1:5:72
time_idx_array = [5, 500, 2500, 5000, 8000, 9998]

idx = 1
for row_id=1:2
    for col_id=1:3
        ax = axes[row_id, col_id]
        time_idx = time_idx_array[idx]
        alpha_beta = (LQDTexpLQDT .* alpha_mat[:, time_idx-1]) .* y_beta_mat[:, time_idx]
        ax.plot(xarray, alpha_beta, "b.", label=@sprintf "t=%d" time_idx)
        ax.vlines(xarray, ymin=0, ymax=alpha_beta, label=@sprintf "sum=%.1E" sum(alpha_beta))
        ax.set_xticks(xticks)
        ax.legend()
        if col_id == 1
            ax.set_ylabel("\$-\\lambda'_i \\Delta t e^{-D \\lambda'_i \\Delta t} \\left< \\hat{\\alpha}_{t-1} | \\psi_i \\right> \\left< \\psi_i | \\mathbf{y}_t | \\hat{\\beta}_{t} \\right>\$")
        end
        idx += 1
    end
end

plt.tight_layout()
#plt.savefig("/home/ytcdata/em_theory/Pictures/ch2/LQ_edt_alpha_beta_in_dldD.pdf")
plt.show()
```

### Part 13-1: $\sum_i e^{-D \lambda'_i \Delta t} \left< \hat{\alpha}_{t-1} | \psi_i \right> \left< \psi_i | \textbf{y}_t | \hat{\beta}_{t} \right>$ vs $t$

```julia
denominator_array = zeros(tau)
alpha_t0 = zeros(Nv)
alpha_t0[1] = 1
denominator_array[1] = dot(expLQDT .* alpha_t0, y_beta_mat[:, 1])
for time_idx=2:tau
    denominator_array[time_idx] = dot(expLQDT .* alpha_mat[:,time_idx-1], y_beta_mat[:, time_idx])
end
```

### Part 13-2: $\sum_i -\lambda'_i \Delta te^{-D \lambda'_i \Delta t} \left< \hat{\alpha}_{t-1} | \psi_i \right> \left< \psi_i | \textbf{y}_t | \hat{\beta}_{t} \right>$ vs $t$

```julia
numerator_array = zeros(tau)
alpha_t0 = zeros(Nv)
alpha_t0[1] = 1
numerator_array[1] = dot(LQDTexpLQDT .* alpha_t0, y_beta_mat[:, 1])
for time_idx=2:tau
    numerator_array[time_idx] = dot(LQDTexpLQDT .* alpha_mat[:,time_idx-1], y_beta_mat[:, time_idx])
end
```

```julia
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(16,4), sharey=true)

xarray = 1:25:10000
ax.plot(xarray, numerator_array[xarray], "b.")
ax.vlines(xarray, ymin=0, ymax=numerator_array[xarray])
ax.set_xlim(-20,10020)
ax.set_ylabel("\$\\sum_i -\\lambda'_i \\Delta t e^{-D \\lambda'_i \\Delta t} \\left< \\hat{\\alpha}_{t-1} | \\psi_i \\right> \\left< \\psi_i | \\mathbf{y}_t | \\hat{\\beta}_{t} \\right>\$")
ax.set_xlabel("t")

plt.tight_layout()
#plt.savefig("/home/yizaochen/Desktop/drawzone_temp/numerator.svg")
plt.show()
```

### Part 13-3: $\frac{\sum_i -\lambda'_i \Delta te^{-D \lambda'_i \Delta t} \left< \hat{\alpha}_{t-1} | \psi_i \right> \left< \psi_i | \textbf{y}_t | \hat{\beta}_{t} \right>}{\sum_i e^{-D \lambda'_i \Delta t} \left< \hat{\alpha}_{t-1} | \psi_i \right> \left< \psi_i | \textbf{y}_t | \hat{\beta}_{t} \right>}$ vs $t$
