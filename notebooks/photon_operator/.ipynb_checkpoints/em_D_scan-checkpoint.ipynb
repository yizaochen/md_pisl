```julia
using FEM, Potential, AlphaBeta, PyPlot, Printf, PhotonOperator, JLD, LinearAlgebra
```

### Part 1: Get collocation points and integration kernel for whole domain

```julia
Nh = 64    # The number of Spectral element
Np = 4     # The order of polynomial which used to interpolate and integration
Nv = 72    # Number of eigenvectors

# Define Physical Domain
xratio = 64. # unit: angstrom
xavg = 0.    # unit: angstrom

x, w, Ldx, L = getLagrange(Np, xratio/Nh)
e_norm = x[end] - x[1]
interpo_xs = x .+ x[end]

N, xref, w0, Ldx, w = get_fem_xref_weights_basis(Nh, Np, xratio, xavg);
```

### Part 2: Set $\sigma$ and $D$ for underlying $p_{\rm eq}$

```julia
D = 500; # Diffusion coefficient
sigma_kref = 8.

kref = get_k_by_sigma(sigma_kref)
Vref = harmonic_well_k_mean(xref, kref, xavg)
rho_eq = get_rhoeq(Vref, w0)
p_eq = rho_eq .* rho_eq;
```

```julia
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(8,4))

ax = axes[1]
ax.plot(xref, Vref, linewidth=2)
ax.set_xlabel("\$x\$", fontsize=12)
ax.set_ylabel("\$V(x)\$", fontsize=12)
ax.set_xticks(-64:16:64)

ax = axes[2]
ax.plot(xref, p_eq, linewidth=2)
ax.set_xlabel("\$x\$", fontsize=12)
ax.set_ylabel("\$p_{\\rm eq}(x)\$", fontsize=12)
ax.set_xticks(-64:16:64)

plt.tight_layout()
plt.show()
```

### Part 3: Set $\sigma$ and $D$ for photon operator(Dirac-Delta) $p_{\rm Dirac}$

```julia
sigma_dirac = 0.5
k_dirac = get_k_by_sigma(sigma_dirac)
V_dirac = harmonic_well_k_mean(xref, k_dirac, xavg)
rho_dirac = get_rhoeq(V_dirac, w0)
p_dirac = rho_dirac .* rho_dirac;
```

```julia
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(12,4))

ax = axes[1]
ax.plot(xref, V_dirac, linewidth=2)
ax.set_xlabel("\$x\$", fontsize=12)
ax.set_ylabel("\$V(x)\$", fontsize=12)
ax.set_xticks(-64:16:64)

ax = axes[2]
ax.plot(xref, p_dirac, linewidth=2)
ax.set_xlabel("\$x\$", fontsize=12)
ax.set_ylabel("\$p_{\\rm Dirac}(x)\$", fontsize=12)
ax.set_xticks(-64:16:64)

plt.tight_layout()
plt.show()
```

### Part 5: Scan Log-likelihood when change D

```julia
sigma_array = [0.5, 0.8, 1.0, 1.5, 2.0, 2.5]

D_array_1 = 100:100:400
D_array_2 = 410:10:600
D_array_3 = 700:100:1500
D_array = float(vcat(D_array_1, D_array_2, D_array_3))
#D_array = [400., 500.]

l_container = Dict([(0, zeros(length(D_array))), 
                    (1, zeros(length(D_array))), 
                    (2, zeros(length(D_array))), 
                    (3, zeros(length(D_array))), 
                    (4, zeros(length(D_array))),
                    (5, zeros(length(D_array)))]);
```

```julia
sigma_idx = 0
for sigma_test in sigma_array
    k_photon = get_k_by_sigma(sigma_test)
    l_array = zeros(length(D_array))
    d_idx = 1
    for D_test in D_array
        l_array[d_idx] = get_loglikelihood(Nh, Np, xratio, xavg, p_eq, D_test, Nv, tau, y_record, save_freq, k_photon)
        d_idx += 1
    end
    l_container[sigma_idx] = l_array
    sigma_idx += 1  
end
```

```julia
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12,4))

sigma_idx = 0
for sigma_test in sigma_array
    label = @sprintf "\$\\sigma=%1.1f\$" sigma_test
    ax.plot(D_array, l_container[sigma_idx], "-o", alpha=0.5, label=label)
    sigma_idx += 1 
end
ax.set_ylabel("log-likelihood")
ax.set_xlabel("D")
ax.axvline(500, color="grey", alpha=0.4)
ax.legend()
title = @sprintf "\$D=%.1f~~~\\Delta t=%.1e~~~T=%.1f~\$" D save_freq total_times
ax.set_title(title)
ax.set_xticks(100:100:1500)

plt.tight_layout()
#plt.savefig(@sprintf "/home/yizaochen/Desktop/drawzone_temp/0/D_scan_0.svg")
plt.show()
```
